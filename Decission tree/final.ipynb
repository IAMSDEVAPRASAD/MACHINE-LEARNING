```python
import tensorflow_decision_forests as tfdf
import pandas
```

```python
path = "/home/dell/Jupyter notebook datas/csv files/archive/Chernobyl_ Chemical_Radiation.csv"
train_df = pandas.read_csv(path, nrows=77,usecols=[2,6,7,8])
```

```python
train_df
```

```python
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(train_df, label="Location")
```

```python
model = tfdf.keras.RandomForestModel()
model.fit(train_ds)
```

```python
tfdf.model_plotter.plot_model_in_colab(model, tree_idx=0)
```

```python
test_path ="/home/dell/Jupyter notebook datas/csv files/archive/Chernobyl_ Chemical_Radiation _test.csv"
# Load the testing dataset
test_df = pandas.read_csv(test_path,usecols=[2,6,7,8])
```

```python
test_df
```

```python
# Convert it to a TensorFlow dataset
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test_df, label="Location")

# Evaluate the model
model.compile(metrics=["accuracy"])
```

```python
print(model.evaluate(train_ds))
print(model.evaluate(test_ds))
```
